```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
```

```python
# create a line
plt.plot(x,y)
```

```python
# Add title and labels
plt.title("Simple Line Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot(x,y)
plt.show()
```

```python
# customising the line plt
plt.plot(x,y,color = "red", marker ="o", linestyle= "--")
```

```python
# Adding lables and title
plt.plot(x,y,color = "green", marker ="o", linestyle= "--")
plt.title("Customized Line Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# creating a multiple plots
y2 = [1,4,6,8,10]
#create a figure
plt.figure(figsize=(10,5))
#create a first subplot
plt.subplot(1,2,1) # 1 row, 2 columns, first subplot
plt.plot(x,y,color = "blue")
plt.title("first plot")
```

```python
# create a second subplot
plt.subplot(1,2,2)
plt.plot(x,y2,color = "Green")
plt.title("second plot")
plt.tight_layout()
```

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
```

```python
plt.scatter(x,y,color = "purple",marker ="x")
plt.title("scatter plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# Bar plot
categories = ['A','B','C','D','E']
values = [5,7,3,4,6]
```

```python
# create a bar plot
plt.bar(categories,values,color = "orange")
plt.title("bar plot")
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [2,3,5,7,11]
y2 = [1,4,6,8,10]
```

```python
#create a line plot
plt.plot(x,y1,label = "Data1",color = "blue")
plt.plot(x,y2,label = "Data2",color = "Green")
# Add title,label and legend
plt.title("Line Plot with legends")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```

```python
# Annotations
plt.plot(x,y)
plt.annotate('Highest Point', xy = (5,11),xytext=(4,12), arrowprops = dict(facecolor = "black",arrowstyle = '->'))
```

```python
# Customizing Axes
plt.plot(x,y)
# set x and y limits
plt.xlim(0,6)
plt.ylim(0,12)
# set custom ticks
plt.xticks([1,2,3,4,5],['One','Two','Three','Four','Five'])
plt.yticks([0,5,10,15],['Zero','Five','Ten','Fifteen'])
```
